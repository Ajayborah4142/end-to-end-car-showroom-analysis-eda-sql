### 3)  Data Distribution & Detect And Handle Outliers 

#### Cars Table

```python

numberic_cols = cars[["Year","Price","Quantity_In_Stock"]]
```

```python
for cols in numberic_cols:
    plt.figure(figsize=(8,6))
    sns.histplot(cars[cols],bins= 30,kde= True)
```

```python
for cols in numberic_cols:
    plt.figure(figsize=(8,6))
    sns.boxplot(cars[cols])
```

#### Customers Table 


```python
plt.figure(figsize=(8,6))
sns.histplot(customers['Age'],bins=20, kde= True)
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(customers['Age'])
```

#### Sales

```python
numberic_columns = sales[['Quantity','Sale_Price']]
```

```python
for columns in numberic_columns:
    plt.figure(figsize= (8,6))
    sns.histplot(sales[columns], kde= True)
```

```python
for columns in numberic_columns:
    plt.figure(figsize= (8,6))
    sns.boxenplot(sales[columns])
```
